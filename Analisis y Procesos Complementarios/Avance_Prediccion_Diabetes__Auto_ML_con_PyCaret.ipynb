###Revisando variables categóricas

```python
for columna in df_diabetes_sin_duplicados.select_dtypes(include='object'):  # Seleccionar columnas categóricas
    etiquetas = df_diabetes_sin_duplicados[columna].unique()  # Obtener etiquetas únicas
    print(f"Etiquetas únicas para la columna '{columna}': {etiquetas}")
```

```python
conteo_etiquetas_gender = df_diabetes_sin_duplicados['gender'].value_counts()
print(conteo_etiquetas_gender)
```

```python
conteo_etiquetas_smoking = df_diabetes_sin_duplicados['smoking_history'].value_counts()
print(conteo_etiquetas_smoking)
```

```python
df_diabetes_filtrado = df_diabetes_sin_duplicados[df_diabetes_sin_duplicados['gender'] != 'Other']
```

```python
df_diabetes_filtrado.head()
```

###Encodeando Variables categóricas

```python
df_encoded = pd.get_dummies(df_diabetes_filtrado, columns=['gender', 'smoking_history'])
```

```python
df_encoded.head()
```

#Separando la data para el pycaret y para la prueba en producción

```python
data = df_encoded.sample(frac=0.95, random_state=768)
```

```python
data
```

```python
data_useen = df_encoded.drop(data.index)#5%
```

```python
data_useen
```

```python
data.reset_index(inplace= True, drop= True)#entrenar y validar
data_useen.reset_index(inplace= True, drop= True)#predecir
```

##Paso 2

```python
from pycaret.classification import setup # seteandolos datos a pycaret
```

```python
model_setup = setup(data=data, target='diabetes', session_id=123)##=radom_state
```
